```python
import pandas as pd
from textblob import TextBlob
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
data= pd.read_csv('full_clean_topics.csv')
```

```python
data['polarity']=0
data["subjectivity"]=0
```

```python
data.head()
```

```python
data.topic.where(data.topic==19).dropna()
```

```python
for i in range(len(data)):
#     print(data.iloc[i,2])
    try:
        blob=TextBlob(data.iloc[i,2])
        result=blob.sentiment        
        data.iloc[i,-2]=result.polarity
        data.iloc[i,-1]=result.subjectivity
    except:
        pass
```

```python
sentiment_raw = pd.pivot_table(data, values='polarity',columns=['topic']).mean()
```

```python
data.loc[data['polarity']<-0.3,'polarity'] =-1
data.loc[(data['polarity']>=0-.3) & (data['polarity']<=0.3),'polarity']=0
data.loc[data['polarity']>0.3,'polarity']=1
```

```python
data.polarity.unique()
```

```python
data.isnull().sum().sum()
```

```python
X=np.array(data['text_final'])
Y=np.array(data['polarity'])
```

```python
X_train=X[:int(.75*len(X))]
Y_train= Y[:int(.75*len(Y))]
X_test=X[int(.75*len(X)):]
Y_test=Y[int(.75*len(Y)):]
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.model_selection import cross_val_score
 
# start with the classic
# with either pure counts or tfidf features

svc = Pipeline([
        ("tfidf_vectorizer", TfidfVectorizer(stop_words="english", max_features=3000)),
        ("linear svc", SVC(C=2,kernel="linear",probability=True,tol=1e-3,cache_size=600,verbose=True,decision_function_shape='ovo'))
    ])
```

```python
model = svc
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)
from sklearn.metrics import classification_report, accuracy_score
print(accuracy_score(Y_test, y_pred))
print(classification_report(Y_test, y_pred))
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, y_pred)
```

```python
r_square_train=model.score(X_train,Y_train)
r_square_train
```

```python
r_square_test=model.score(X_test,Y_test)
r_square_test
```

```python
topics=data.topic.unique()
topics.sort()
polarities=data.polarity.unique()
new=pd.DataFrame()
new['positive']=np.nan
new['negative']=np.nan
new['zero']=np.nan
for i in (topics):
    i=int(i)
    new.loc[i,'positive']=data.topic.where((data.polarity==1)&(data.topic==i)).count()
    new.loc[i,'negative']=data.topic.where((data.polarity==-1)&(data.topic==i)).count()
    new.loc[i,'zero']=data.topic.where((data.polarity==0)&(data.topic==i)).count()
labels = topics
positive = new['positive']
negative = new['negative']
zero=new['zero']
x = np.arange(len(labels))  # the label locations
width = 0.3  # the width of the bars

fig, ax = plt.subplots()
rects3 = ax.bar(x , zero, width, label='zero',color='b')
rects1 = ax.bar(x - .3, positive, width, label='positive',color='g')
rects2 = ax.bar(x + .3, negative, width, label='negative',color='r')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
#fig.tight_layout()

plt.show()
```
