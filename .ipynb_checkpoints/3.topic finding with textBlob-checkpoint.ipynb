```python
import pandas as pd
from textblob import TextBlob
import numpy as np
```

```python
data=pd.read_csv("full_clean.csv")
df=pd.DataFrame(data.text_final)
df.dropna(inplace=True)
topics=pd.read_csv("topics_full.csv")
topics=pd.DataFrame(topics)
topics.dropna(inplace=True)
# topic_lda['topic']=topics.lda
# topic_nmf['topic']=topics.nmf
```

```python
wordscount= pd.read_csv('wordscount.csv')
wordscount['blob']= [TextBlob(entry) for entry in wordscount['words']]
wordscount['blob'] = [entry.words for entry in wordscount['blob']]
```

```python
# lda=pd.DataFrame(topics['lda'])
```

```python
topics['topic']=topics['nmf']
```

```python
topics.drop(['Unnamed: 0','lda','nmf'],1,inplace=True)
```

```python
df['blob']= [TextBlob(entry) for entry in df['text_final']]
topics['blob']= [TextBlob(entry) for entry in topics['topic']]
```

```python
df['blob'] = [entry.replace("'","") for entry in df['blob']]
```

```python
df['blob'] = [entry.words for entry in df['blob']]
topics['blob'] = [entry.words for entry in topics['blob']]
```

```python
df["topic"]=0
```

```python
topics.head()
```

```python
for dfline in range(len(df)):
    dflist=[]
    for dfword in df.iloc[dfline,1]:
        dflist.append(dfword)
    for topicline in range(len(topics)):
        for topicword in topics.iloc[topicline,1]:
            if (topicword) in (dflist):
                df.iloc[dfline,2]=topicline
                continue
```

```python
df.topic.where(df.topic==19).dropna().count()
```

```python
data['topic']=df['topic']
```

```python
for lines in range(len(topics)):
    numbers=[]
    for words in topics.loc[lines,'blob']:
        for line in range(len(wordscount)):
            for word in wordscount.loc[line,'blob']:
                if words==word:
                    numbers.append(wordscount.loc[line,'counts'])
    topics.loc[lines,'count']= (numbers)#agar error dad, aval str(numbers), baad (numbers)
```

```python
for line in range(len(topics)):
    j=0
    sums=0
    darsad=[]
    for number in topics.loc[line,'count']:
        sums=int(number)+(sums)
    for number in topics.loc[line,'count']:
        darsad.append(100*(number/sums))
    topics.loc[line,'darsad']=str(darsad)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(16,12)
colors=['C1','C2','C3','C4','C5','C6','C7','C8']

for line in range(len(topics)):
    try:
        text=np.array(topics.loc[line,'blob'])
        numbers=np.array(topics.loc[line,'count'])

        plt.bar(text,numbers,color=colors)
        plt.title('topic number '+str(line))
        plt.xticks(rotation=45)
        plt.show()
    except:
        pass
```
